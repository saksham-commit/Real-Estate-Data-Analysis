```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("real_estate_data.csv")
```

```python
print(df.isnull().sum())
```

```python
df.dtypes
```

```python
df.head(30)
```

```python
columns_with_missing_values = ["Bedrooms","Bathrooms","Price_Lakhs", "Area_sq_ft"]  

for col in columns_with_missing_values:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
df.head(30)
```

```python
columns_to_convert = ["Bedrooms","Bathrooms","Price_Lakhs", "Area_sq_ft"]  # Add more if needed

for col in columns_to_convert:
    df[col] = df[col].astype(int)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12,10))
ax=sns.barplot(x="City" ,y="Price_Lakhs",data=df,ci=None)
plt.title("Distribution of Property Prices (in Lakhs)")
plt.xlabel("Locations of Properties")
plt.ylabel("prices of properties")
ax.bar_label(ax.containers[0])
plt.show()
```

The above graph shows that chennai has more properties as compared to other and the average value is 297.219 (in lakhs).

```python
plt.figure(figsize=(12,10))
ax=sns.barplot(x="Property_Type" ,y="Price_Lakhs",data=df,ci=None,palette='pastel')
plt.title("Distribution of Property Type with Price")
plt.xlabel("Property_Type")
plt.ylabel("prices of properties")
ax.bar_label(ax.containers[0])
plt.show()
```

It shows that the customer acquire more apartments and the average value of apartments is 294.249 (in lakhs).

```python
plt.figure(figsize=(14, 15))
cp=sns.countplot(x="Property_Type", hue="City", data=df, palette="viridis")
plt.title("Total Properties in Different Locations")
plt.xlabel("Property Type")
plt.ylabel("Count")
for container in cp.containers:
    cp.bar_label(container)
plt.show()
```

The graph shows that Mumbai acquire more apartments whose count is 33 and delhi acquire more villa whose count is 31.

```python
plt.figure(figsize=(10,8))
cp1=sns.countplot(x="Furnishing_Status",data=df)
plt.title(" Furnishing Status of Property")
cp1.bar_label(cp1.containers[0])
plt.xlabel("Furnishing Status")
plt.show()
```

It shows that the customer acquire more unfurnished property and less number of furnished pproperty was acquired.

```python
plt.figure(figsize=(8, 5))
hp=sns.histplot(x="Price_Lakhs",data=df, color='blue',bins=5)
plt.title("Distribution of Property Prices (in Lakhs)")
plt.xlabel("Price in Lakhs")
plt.ylabel("Frequency")
hp.bar_label(hp.containers[0])
plt.show()
```

The graph shows that the most properties are lies between the price of 200-300 (in Lakhs).
